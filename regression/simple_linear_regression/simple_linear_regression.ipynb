```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =pd.read_csv("Salary_Data.csv")
print(dataset.head())
X=dataset.iloc[:,0].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
_test_size = 1/3
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=_test_size, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
```

```python
predicted_y = regressor.predict(X_test.reshape(-1, 1)) 
```

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train.reshape(-1, 1)), color = "blue")
plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test, y_test, color = "red")
#  mantidado pq a linha de regreo do teste  a mesma do treino
plt.plot(X_train, regressor.predict(X_train.reshape(-1, 1)), color = "blue")
plt.title("Salary vs Experience (Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# predicting a single salary
regressor.predict([[4]])
```

```python
# getting coeficientes
print(regressor.coef_)
print(regressor.intercept_)

print("Salary = {} * YearsExperience + {}".format(regressor.coef_[0][0],regressor.intercept_[0]))
```
