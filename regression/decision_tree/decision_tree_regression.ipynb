É interessante para muitos dados e com alta dimensão

- faz essas div ótima baseados na entropia
Então essas div (áreas) conseguem se formadas atráves de uma arvore de decisão



Se um valor cair aqui, então ele será generalizado para o valor de y da médias dos valores dessa mesma folha

Baseada na média das folhas, temos:



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0) #apenas para obter o mesmo result
regressor.fit(X,y)
```

```python
regressor.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = "red")
plt.plot(X_grid, regressor.predict(X_grid), color ="blue")
plt.title("Truth or Bluff (Decision Tree Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
